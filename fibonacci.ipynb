## ***Lgica Bsica***

````python
a = 0 
b = 1 
for i in range(8):
    print(b)            # na 1 iterao "B" tem o valor de 1, na 2 iterao ser igual ao valor do "C"    
    c = a + b           # na 1 iterao "A" tem o valor de 0, na 2 ser igual ao valor do 1 "B"
    a = b               # a cada interao "A" valer igual ao "B" do print
    b = c               # a cada iterao "B" valer o "C"
# resultado:   1, 1, 2, 3, 5, 8, 13, 21
````

````python
a = 0
b = 1
for i in range(8):
    print(f'a: {a} | b: {b} | c(a+b): {c} | Resultado (b): {b}')
    c = a + b
    a = b
    b = c
````

## ***Lgica otimizada***

````python
a, b = 0, 1
for i in range(8):
    print(b, end=', ')
    a, b = b, a + b
# resultado:  1, 1, 2, 3, 5, 8, 13, 21
````

## ***Lgica bsica utilizando LISTA***

````python
lista = []
a = 0
b = 1
for i in range(8):
    lista.append(b) 
    c = a + b
    a = b  
    b = c  
print(*lista, sep = ', ')
# resultado:  1, 1, 2, 3, 5, 8, 13, 21
````

## ***Lgica Otimizada e usando LISTA***

````python
a, b = 0, 1
lista = []
for i in range(8):
    lista.append(b)
    a, b = b, a + b
print(*lista, sep = ', ')
````
