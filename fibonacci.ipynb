## ***Lógica Básica***

````python
a = 0 
b = 1 
for i in range(8):
    print(b)            # na 1ª iteração "B" tem o valor de 1, na 2º iteração será igual ao valor do "C"    
    c = a + b           # na 1ª iteração "A" tem o valor de 0, na 2º será igual ao valor do 1ª "B"
    a = b               # a cada interação "A" valerá igual ao "B" do print
    b = c               # a cada iteração "B" valerá o "C"
# resultado:   1, 1, 2, 3, 5, 8, 13, 21
````

````python
a = 0
b = 1
for i in range(8):
    print(f'a: {a} | b: {b} | c(a+b): {c} | Resultado (b): {b}')
    c = a + b
    a = b
    b = c
````

## ***Lógica otimizada***

````python
a, b = 0, 1
for i in range(8):
    print(b, end=', ')
    a, b = b, a + b
# resultado:  1, 1, 2, 3, 5, 8, 13, 21
````

## ***Lógica básica utilizando LISTA***

````python
lista = []
a = 0
b = 1
for i in range(8):
    lista.append(b) 
    c = a + b
    a = b  
    b = c  
print(*lista, sep = ', ')
# resultado:  1, 1, 2, 3, 5, 8, 13, 21
````

## ***Lógica Otimizada e usando LISTA***

````python
a, b = 0, 1
lista = []
for i in range(8):
    lista.append(b)
    a, b = b, a + b
print(*lista, sep = ', ')
````
